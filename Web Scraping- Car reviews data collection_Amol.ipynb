```python
#Creating the dataframe with the column names 
df = pd.DataFrame(columns=["review text","review date","review star-rating","reviewer"])
```

```python
#creating empty list for each filed in which data will be appended using for loop
Review_list =[]
Reviewdate_list = []
Rating_list = []
Reviewer_name_list = []
```

```python
#Creating function to extract Reviews data
def Review_fn(titles):
    for title in titles:
        User_Review = title.text
        Review_list.append(User_Review)
        print(User_Review)
```

```python
#Creating function to extract review date
def Date_fn(dates):
    for date in dates:
        Review_Date = date.text
        Reviewdate_list.append(Review_Date)
        print(Review_Date)
```

```python
#Creating function to extract rating data
def Rating_fn(ratings):
    for rating in ratings:
        User_Rating= rating.text
        Rating_list.append(User_Rating)
        print(User_Rating)
```

```python
#Creating function to extract reviewer name data
def Reviewer_fn(reviewers):
    for reviewer in reviewers:
        Reviewer_nm= reviewer.text
        Reviewer_name_list.append(Reviewer_nm)
        print (Reviewer_nm)
```

```python
csv_file = open('jeep_compass.csv', 'w', encoding="UTF-8", newline="")
writer = csv.writer(csv_file)

# Dataframe column name to be stored 
writer.writerow(
    ['review_title', 'review_url', 'review_text', 'review_likes', 'review_rating'])

for i in range(2,101,1):
    url = "https://www.cardekho.com/jeep/compass/user-reviews/"+str(i)
    driver.get(url)
    Reviews = driver.find_elements_by_class_name('readReviewHolder')
    #print(Reviews)
    print(len(Reviews))
    for rvw in range(1,len(Reviews)):
        review_dict = {}
        try:
            review_title=Reviews[rvw].find_element_by_tag_name('h3').text
        except:
            review_title=""
        try:
            review_url=driver.current_url
        except:
            review_url=""
        try:
            review_text=Reviews[rvw].find_element_by_class_name('contentheight').text
        except:
            review_text=""
        try:
            if "Read More" not in Reviews[rvw].find_element_by_class_name('hover').text:
                review_likes=Reviews[rvw].find_element_by_class_name('hover').text                
            else:
                review_likes            
        except:
            review_likes=""
        try: 
            review_rating=len(Reviews[rvw].find_elements_by_class_name('icon-star-full-fill'))
        except:
            review_rating=""
        review_dict['review_title'] = review_title
        review_dict['review_url']=review_url
        review_dict['review_text']=review_text
        review_dict['review_likes']=review_likes
        review_dict['review_rating']=review_rating
        writer.writerow(review_dict.values())
        
```

```python
#creating try except block for extraction of data from HTML classes using pagination by automatically clicking next button by chrome driver 
#url = "https://www.cardekho.com/kia/seltos/user-reviews"
#url = "https://www.cardekho.com/mg/hector/user-reviews"
url = "https://www.cardekho.com/jeep/compass/user-reviews"
driver.get(url)
titles = driver.find_elements_by_class_name("Next Page")
Review_fn(titles)
for i in range(1,25,1):
    #url = "https://www.cardekho.com/kia/seltos/user-reviews/"+str(i)+"?subtab=latest"
    #url = "https://www.cardekho.com/mg/hector/user-reviews/"+str(i)
    url = "https://www.cardekho.com/jeep/compass/user-reviews/"+str(i)
    driver.get(url)
    buttons = driver.find_elements_by_class_name('Next Page')
    for button in buttons:
        try:
            button.click()
        except:
            pass
    titles = driver.find_elements_by_xpath('//div//h3//a')
    Review_fn(titles)
    dates = driver.find_elements_by_xpath('//div[@class="authorSummary"]//div[@class="date"]')
    Date_fn(dates)
    ratings = driver.find_elements_by_class_name('rating')
    Rating_fn(ratings)
    reviewers = driver.find_elements_by_class_name('name')
    Reviewer_fn(reviewers)

# Close the files
file.close()
driver.close()
```

```python
#Creating dataframe with the appended list of required fields
df = pd.DataFrame(list(zip(Review_list, Reviewdate_list, Rating_list, Reviewer_name_list)), columns=['review text','review date','review star-rating','reviewer'])
```
