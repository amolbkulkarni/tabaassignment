```python
#Importing the  necessary python libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from math import ceil
import pandas as pd
import time
import os
import csv
import pandas as pd
from math import ceil
```

```python
# Providing location of chrome driver exe file 
locationOfWebdriver = "C:/Users/Amol/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(locationOfWebdriver)
```

```python
csv_file = open('F://ISB//TABA//new_car_reviews.csv', 'w', encoding="UTF-8", newline="")
writer = csv.writer(csv_file)

# Dataframe column name to be stored 
writer.writerow(
    ['review_title', 'review_url', 'review_text', 'review_likes', 'review_rating'])

urls = ["https://www.cardekho.com/jeep/compass/user-reviews/","https://www.cardekho.com/kia/seltos/user-reviews/","https://www.cardekho.com/mg/hector/user-reviews/"]
for urll in urls:
        for i in range(2,101,1):
            url = urll+str(i)
            driver.get(url)
            Reviews = driver.find_elements_by_class_name('readReviewHolder')
            #print(Reviews)
            print(len(Reviews))
            for rvw in range(1,len(Reviews)):
                review_dict = {}
                try:
                    review_title=Reviews[rvw].find_element_by_tag_name('h3').text
                except:
                    review_title=""
                try:
                    review_url=driver.current_url
                except:
                    review_url=""
                try:
                    review_text=Reviews[rvw].find_element_by_class_name('contentheight').text
                except:
                    review_text=""
                try:
                    if "Read More" not in Reviews[rvw].find_element_by_class_name('hover').text:
                        review_likes=Reviews[rvw].find_element_by_class_name('hover').text                
                    else:
                        review_likes            
                except:
                    review_likes=""
                try: 
                    review_rating=len(Reviews[rvw].find_elements_by_class_name('icon-star-full-fill'))
                except:
                    review_rating=""
                review_dict['review_title'] = review_title
                review_dict['review_url']=review_url
                review_dict['review_text']=review_text
                review_dict['review_likes']=review_likes
                review_dict['review_rating']=review_rating
                writer.writerow(review_dict.values())
```
